```python
#Description:Operational lifecycle tracking of orders.
# Grain:One row per order
```

```python
orders_silver_df=spark.read.table("git_analysis.silver.orders")
```

```python
from pyspark.sql import functions as F
gold_order_funnel_df = (
    orders_silver_df
    .groupBy("order_id")
    .agg(
        F.min("order_placed_date").alias("order_created_ts"),
        F.max("update_date").alias("last_update_ts"),
        F.first("order_status").alias("final_status")
    )
)
```

```python
gold_order_funnel_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("git_analysis.gold.gold_order_funnel")
```
