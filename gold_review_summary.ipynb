```python
#Description:Product sentiment tracking
#Product-level OR product-month
#Metrics:avg_rating,rating_distribution,review_volume
```

```python
reviews_silver_df=spark.read.table("git_analysis.silver.reviews")
```

```python
from pyspark.sql import functions as F

gold_review_summary_df = (
    reviews_silver_df
    .groupBy("product_id")
    .agg(
        F.count("review_id").alias("review_count"),
        F.avg("rating").alias("avg_rating"),
        F.min("review_timestamp").alias("first_review_date"),
        F.max("review_timestamp").alias("last_review_date"),

        # Rating distribution
        F.sum(F.when(F.col("rating") == 5, 1).otherwise(0)).alias("rating_5_count"),
        F.sum(F.when(F.col("rating") == 4, 1).otherwise(0)).alias("rating_4_count"),
        F.sum(F.when(F.col("rating") == 3, 1).otherwise(0)).alias("rating_3_count"),
        F.sum(F.when(F.col("rating") == 2, 1).otherwise(0)).alias("rating_2_count"),
        F.sum(F.when(F.col("rating") == 1, 1).otherwise(0)).alias("rating_1_count")
    )
)
```

```python
gold_review_summary_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("git_analysis.gold.gold_review_summary")
#
```
