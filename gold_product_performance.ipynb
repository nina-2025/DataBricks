```python
#Description:Aggregated product sales and review performance.
#Grain:One row per product
```

```python
transactions_silver_df=spark.read.table("git_analysis.silver.transactions")
```

```python
from pyspark.sql import functions as F
gold_product_performance = (
    transactions_silver_df
    .withColumn("order_date", F.to_date("order_timestamp"))
    .groupBy("order_date")
    .agg(
        F.countDistinct("order_id").alias("daily_orders"),
        F.sum("quantity").alias("daily_units_sold"),
        F.sum(F.col("quantity") * F.col("unit_price")).alias("daily_revenue"),
        F.avg(F.col("quantity") * F.col("unit_price")).alias("avg_order_value")
    )
)
```

```python
gold_product_performance.write.format("delta").mode("overwrite").saveAsTable("git_analysis.gold.gold_product_perfomance")
```
