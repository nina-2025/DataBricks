```python

## # **defining schemas**
```

```python
%python
from pyspark.sql.types import StructType, IntegerType, StringType, DateType,StructField
customers_schema = StructType(fields=[StructField("customer_id",StringType(),False),StructField("name",StringType(),True),StructField("region",StringType(), True),StructField("country",StringType(), True),StructField("signup_date",DateType(),True)  ])
```

```python
%python
from pyspark.sql.types import StructType, IntegerType, StringType, DateType,StructField
transactions_schema = StructType(fields=[StructField("order_id",StringType(),False),StructField("customer_id",StringType(),True),StructField("product_id",StringType(), True),StructField("quantity",IntegerType(), True),StructField("unit_price",DoubleType(),True), StructField("order_timestamp",TimestampType(),True),StructField("status",StringType(),True), StructField("salesperson",StringType(),True) ])
```

```python
%python
from pyspark.sql.types import StructType, IntegerType, StringType, DateType,StructField
orderitems_schema = StructType(fields=[StructField("ORDERITEMSID",StringType(),False),StructField("ORDERID",StringType(),True),StructField("PRODUCTID",StringType(), True),StructField("QUANTITY",IntegerType(), True)  ])
```

```python
%python
from pyspark.sql.types import StructType, StringType,IntegerType, TimestampType,StructField
reviews_schema = StructType(fields=[StructField("review_id",StringType(),False),StructField("customer_id",StringType(),True),StructField("product_id",StringType(), True),StructField("rating",IntegerType(), True),StructField("review_text",StringType(), True),StructField("review_timestamp",TimestampType(), True)   ])
```

```python
%python
from pyspark.sql.types import StructType, StringType, DateType,StructField
order_schema = StructType(fields=[StructField("ORDERID",StringType(),False),StructField("CUSTOMERID",StringType(),True),StructField("SALESPERSON",StringType(), True),StructField("ORDERSTATUS",StringType(), True),StructField("ORDERPLACEDTIMESTAMP",DateType(), True),StructField("UPDATEDAT",DateType(), True)   ])
```

```python
%python
from pyspark.sql.types import StructType, StringType, DoubleType
products_schema = StructType(fields=[StructField("product_id",StringType(),False),StructField("product_name",StringType(),True),StructField("category",StringType(), True),StructField("price",DoubleType(), True)   ])

```

## creating tables

```python
%python
customer_df = (
    spark.read
    .schema(customers_schema)
    .csv('/Volumes/git_analysis/bronze/raw/raw_data/raw_customers.csv', header=True)
)

customer_df.write.format("delta").mode("overwrite").saveAsTable(
    "git_analysis.bronze.customers"
)
```

```python
%python
orderitems_df = (
    spark.read
    .schema(orderitems_schema)
    .csv('/Volumes/git_analysis/bronze/raw/raw_data/raw_orderitems.csv', header=True)
)

orderitems_df.write.format("delta").mode("overwrite").saveAsTable(
    "git_analysis.bronze.orderitems"
)


```

```python
%python
order_df = (
    spark.read
    .schema(order_schema)
    .csv('/Volumes/git_analysis/bronze/raw/raw_data/raw_orders.csv', header=True)
)

order_df.write.format("delta").mode("overwrite").saveAsTable(
    "git_analysis.bronze.order"
)


```

```python
%python
products_df = (
    spark.read
    .schema(products_schema)
    .csv('/Volumes/git_analysis/bronze/raw/raw_data/raw_products.csv', header=True)
)

products_df.write.format("delta").mode("overwrite").saveAsTable(
    "git_analysis.bronze.products"
)


```

```python
%python
reviews_df = (
    spark.read
    .schema(reviews_schema)
    .csv('/Volumes/git_analysis/bronze/raw/raw_data/raw_reviews.csv', header=True)
)

reviews_df.write.format("delta").mode("overwrite").saveAsTable(
    "git_analysis.bronze.reviews"
)


```

```python
%python
transactions_df = (
    spark.read
    .schema(transactions_schema)
    .csv('/Volumes/git_analysis/bronze/raw/raw_data/raw_transactions.csv', header=True)
)

transactions_df.write.format("delta").mode("overwrite").saveAsTable(
    "git_analysis.bronze.transactions"
)


```
